# Упражнение 1. Бинарный поиск

Лучший способ искать элемент в отсортированном массиве. Пусть требуется определить индекс $i$ элемента, равного `x`, в упорядоченном по возрастанию массиве $A$. Тогда действовать следует следующим образом:

1. Берем центральный элемент $c$ в массиве $A$

2. Сравниваем $c$ с $x$. Если $c < x$, то элемент, равный $x$ правее $c$ и бинарный поиск применяется к правой половине массива. Если $c < x$, то бинарный поиск соответственно применяется к левой половине массива.

Реализуйте поиск элемента, равного $x$, в рекурсивном и нерекурсивном вариантах.

````python
import random


# получим случайный отсортированный массив с неповторяющимися элементами
A = [random.randint(1, 500) for _ in range(1, 25)]
A.sort()
for i in range(len(A)-1):
    if A[i+1] == A[i]: A[i+1] += 1
        
# ищем элемент x
X = A[random.randint(0, len(A) - 1)]


# начинаем с +/- 1 элемента, чтобы хватало шагов для поиска крайних элементов: A[0], A[len(A) - 1].
def good_binary_search(x, a):
    left = -1 
    right = len(A)
    while right > left + 1: 
        middle = (left + right) // 2 
        if a[middle] >= x: 
            right = middle 
        else: 
            left = middle 
    return right


# здесь все понятно
def biary_search_recursive (x, a):
    if x == a[len(a)//2]:
        return len(a)//2
    elif x < a[len(a)//2]:
        a_new = a[0:len(a)//2]
        return biary_search_recursive (x, a_new)
    if x > a[len(a)//2]:
        a_new = a[(len(a)//2 + 1):(len(a))]
        return len(a)//2 + biary_search_recursive (x, a_new) + 1

    
# main    
print("ищем элемент ", X, "в массиве ", A)
print("его индекс, найденный циклом ", good_binary_search(X, A))
print("его индекс, найденный рекурсией ", biary_search_recursive(X, A))
````

# Упражнение 2. Поиск точки максимума в целых числах

По целому $c \ge 1$ найдите целое $a$, при котором $a^{c-a}$ максимально. Используйте бинарный поиск.

````python
import math
import random


C = random.randint(1, 500)


# ищем максимум фунцкии a^(c-x). он совпадает с максимумом функции ln[a^(c-x)]
# эта функция имеет максимум в точке a + aln(a) = c или ближайшей к ней целой точке.
# ищем эту точку бинарным поиском, тк a + aln(a) строго возрастает

def find_a(c):
    left = -1
    right = c
    while right > left + 1:
        a  = (left + right) // 2
        if a + a*math.log(a) == c:
            return a
        elif a + a*math.log(a) > c:
            right = a
        else:
            left = a
    if (a ** (c-a) > (a+1) ** (c-a-1)) and (a ** (c-a) > (a-1) ** (c-a+1)):
        return a
    if ((a+1) ** (c-a-1) > a ** (c-a)) and ((a+1) ** (c-a-1) > (a-1) ** (c-a+1)):
        return a + 1
    if ((a-1) ** (c-a+1) > a ** (c-a)) and ((a-1) ** (c-a+1) > (a+1) ** (c-a-1)):
        return a - 1
    
# последняя проверка делается на случай ошибки в последнем шаге: может сложится ситуация,    
# при которой модуль a + aln(a) - c для предпоследнего шага индукции меньше того же значения
# для последнего (это может случится, например, если a[i-1] = a[i] + 1)


# main
A = find_a(C)
B = A ** (C - A)
print('при получении на вход C = ', C, 'получили максимум при A = ', A, 'и значение A ** (C-A) = ', B)
````

# Упражнение 3. $k$-й по величине элемент

Найти $k$-е по величине значение в массиве из $N$ элементов (не сортируя массив полностью). Если сами не придумаете, есть подсказка к задаче 5.6 на стр. 438 в [1].

````python
import random


# получим случайный неотсортированный массив с возможно повторяющимися элементами.
A = [random.randint(1, 500) for _ in range(random.randint(1, 25))]
K = random.randint(1, len(A))


# буду считать, что k-е по порядку - имеется ввиду (k-1)-ый элемент в отсортированном
# полностью массиве, если бы его можно было отсортировать. Отсчет ведется от 0-го элемента.
def find_k(k, a):
    k_left = 0
    k_right = len(a) - 1
    m1, m2 = 0, 0
    current_k = k
    
    while True:
        m1, m2 = one_sort(a, k_left, k_right)
        if m1 >= current_k:
            k_right = k_left + m1 - 1
        elif m1 + m2 >= current_k:
            return a[k-1]
        else:
            k_left = k_left + m1 + m2
            current_k = current_k - m1 - m2
    
    
# единичный акт сортировки, left и right - включительно
def one_sort(a, left, right):
    m1, m2, m3 = 0, 0, 0
    reference = a[random.randint(left, right)]
    current_index = left
    
    while m1 + m2 + m3 < right - (left-1) :
        if a[current_index] < reference:
            a[left + m1], a[current_index] = a[current_index], a[left + m1]
            m1 += 1
            current_index += 1
        elif a[current_index] == reference:
            m2 += 1
            current_index += 1
        elif a[current_index] > reference:
            a[current_index], a[right - m3] = a[right - m3], a[current_index]
            m3 += 1  
    return m1, m2


# main
print(K, '-м по возрастанию элементом массива ', A)
T = find_k(K, A)
print('является число ', T)
````

##### Упражнение 4. Разбиение Ломуто

Реализуйте быструю сортировку на основе [разбиения Ломуто](https://en.wikipedia.org/wiki/Quicksort#Lomuto_partition_scheme).

````python
import random


# получим случайный неотсортированный массив с возможно повторяющимися элементами.
A = [random.randint(1, 500) for _ in range(random.randint(1, 25))]


# один проход сортировки, left и right включены в массив. На вход подавать массивы
# содержащие больше двух элементов. Переделано из one_sort()       
def Lomuto_partition(a, left, right):
    m1, m2, m3 = 0, 0, 0
    reference = a[right]
    current_index = left
    
    while m1 + m2 + m3 < right - (left-1) :
        if a[current_index] < reference:
            a[left + m1], a[current_index] = a[current_index], a[left + m1]
            m1 += 1
            current_index += 1
        elif a[current_index] == reference:
            m2 += 1
            current_index += 1
        elif a[current_index] > reference:
            a[current_index], a[right - m3] = a[right - m3], a[current_index]
            m3 += 1
    return m1, m3

                
def Lomuto_sort(a, left, right):
    m1, m3 = Lomuto_partition(a, left, right)
    if m1 >= 2:
        Lomuto_sort(a, left, left + m1 - 1)
    if m3 >= 2:
        Lomuto_sort(a, right - m3 + 1, right)


# main        
print('подав на вход массив ', A)
Lomuto_sort(A, 0, len(A) - 1)
print('получили на выходе остортированный массив', A)
````

# Упражнение 5. Задача флага Нидерландов

Пусть есть массив, элементы которого равны одному из чисел 0, 1, 2. Как упорядочить такой массив возрастанию в один проход?<br>
Попробуйте придумать решение самостоятельно, а затем сравните с алгоритмом, описанном [здесь](https://neerc.ifmo.ru/wiki/index.php?title=Dutch_national_flag_problem). Реализуйте быструю сортировку на основе описанного разбиения. Функция `partition()` должна разбивать массив на 3 части: в первой должны быть элементы меньше опорного, во второй - равные опорному, в третей - элементы больше опорного.

> [Здесь](https://algs4.cs.princeton.edu/lectures/23DemoPartitioning.pdf) есть демонстрация работы разбиения под названием "Dijkstra 3-way partitioning" 

> Запись `++` может обозначать либо префиксный, либо постфиксный инкремент, в зависимости от того, где расположен операнд. Инструкция
```c++
a = ++i
```
содержит оператор "префиксный инкремент", который сначала увеличивает `i` на 1, а затем возвращает значение `i`. Например, если `i == 0`, то после выполнения инструкции `i == 1`, `a == 1`. В следующем фрагменте показан постфиксный инкремент.
```c++
a = i++
```
В этом случае сначала возвращается значение `i` до увеличения, а затем уже выполняется увеличение `i` на 1. Пусть `i == 0`, тогда в конце получится `i == 1`, `a == 0`.

> Перед тем обменивать 

````python
import random


# получим случайный неотсортированный массив из 3-х видов чисел.
A = [random.randint(0, 2) for _ in range(random.randint(1, 25))]


# переделано из one_sort()
def flag_sort(a):
    reference = 1
    current_index = 0
    m1, m2, m3 = 0, 0, 0
    
    while m1 + m2 + m3 < len(a) :
        if a[current_index] < reference:
            a[m1], a[current_index] = a[current_index], a[m1]
            m1 += 1
            current_index += 1
        elif a[current_index] == reference:
            m2 += 1
            current_index += 1
        elif a[current_index] > reference:
            a[current_index], a[len(a) - 1 - m3] = a[len(a) - 1 - m3], a[current_index]
            m3 += 1

            
# main
print('массив ', A, 'отсортирован в один проход.')
flag_sort(A)
print('Получился массив', A)
````

# Упражнение 6. Быстрая сортировка на 2-х опорных элементах (dual pivot quicksort)

Есть улучшенная версия разбиения из упражнения 5. Выбираются **2** различных опорных элемента, и меньший из которых $p_1$ располагается в начале массива, а больший $p_2$ в конце массива. Если такие опорные элементы не удалось выбрать, то массив состоит из одинаковых элементов и уже отсортирован. Далее элементы, находящиеся между опорными распределяются, как в упражнении 5, но в центральную часть попадают элементы, чье значение $v$ удовлетворяет неравенствам $p_1 \le v \le p_2$. После распределения опорные элементы $p_1$ и $p_2$ присоединяются к центральной части соответственно с начала и с конца. 

Реализация на C есть [тут](https://www.geeksforgeeks.org/dual-pivot-quicksort/)

> [Здесь](https://algs4.cs.princeton.edu/lectures/23DemoPartitioning.pdf) есть демонстрация работы разбиения под названием "dual-pivot partitioning" 

````python
import random


# получим случайный неотсортированный массив с возможно повторяющимися элементами.
A = [random.randint(1, 500) for _ in range(random.randint(1, 25))]


# один проход сортировки, left и right включены в массив. На вход подавать массивы
# содержащие больше двух элементов. Переделано из one_sort().
def double_partition(a, left, right):
    m1, m2, m3 = 0, 0, 0

    if a[left] > a[right]:
        a[left], a[right] = a[right], a[left] 
    
    reference_left = a[left]
    reference_right = a[right]
    lt = left + 1
    gt = right - 1
    current_index = left + 1
    
    while not current_index > gt:
        if a[current_index] < reference_left:
            a[lt], a[current_index] = a[current_index], a[lt]
            m1 += 1
            current_index += 1
            lt += 1
        elif a[current_index] > reference_right:
            a[current_index], a[gt] = a[gt], a[current_index]
            m3 += 1
            gt -= 1
        else:
            m2 += 1
            current_index += 1
    lt -= 1
    gt += 1
    a[left], a[lt] = a[lt], a[left]
    a[right], a[gt] = a[gt], a[right]
    return m1, m2, m3


# все тоже самое, только элементы a[left + m1], a[right - m3] оставляем в покое
# тк они уже отсортированны
def double_sort(a, left, right):
    m1, m2, m3 = double_partition(a, left, right)
    
    if m1 >= 2:
        double_sort(a, left, left + m1 - 1)
    if m2 >= 2:
        double_sort(a, left + m1 + 1, right - m3 - 1)
    if m3 >= 2:
        double_sort(a, right - m3 + 1, right)


# main        
print('подав на вход массив ', A)
double_sort(A, 0, len(A) - 1)
print('получили на выходе остортированный массив', A)
````

Два множества целых чисел через пробел записаны в файлах <font color=green>set1.txt</font> и <font color=green>set2.txt</font> в порядке возрастания. Создайте из них третье множество чисел и запишите его в файл в порядке возрастания.

Напишите код для создания третьего множества каждым из перечисленных способов:

1. объединение,

- пересечение,

- разность,

- симметричная разность.

Алгоритм должен быть однопроходным. Не используйте объекты типа `set`.

````python
import random


# получим 2 случайных массива
example_set1 = [random.randint(1, 20) for _ in range(random.randint(1, 25))]
example_set2 = [random.randint(1, 20) for _ in range(random.randint(1, 25))]
# остортируем их
example_set1.sort()
example_set2.sort()
# получим из них кортежи
example_set1 = set(example_set1)
example_set2 = set(example_set2)
# запишем их в файлы
with open('set1.txt', 'w') as f:
    for elem in example_set1:
        f.write(str(elem))
        f.write(' ')
with open('set2.txt', 'w') as f:
    for elem in example_set2:
        f.write(str(elem))
        f.write(' ')

        
def combination(filename1, filename2):
# считаем множества из файлов
    with open(filename1, 'r') as f:
        string = f.read()
    set1 = string.split()
    for i in range(len(set1)):
        set1[i] = int(set1[i])
    with open(filename2, 'r') as f:
        string = f.read()
    set2 = string.split()
    for i in range(len(set2)):
        set2[i] = int(set2[i])
    
    print('множество 1: ', set1)
    print('множество 2: ', set2)
    
    i, j = 0, 0
    set3 = []
    while i < len(set1) or j < len(set2):
        if j == len(set2) or (i < len(set1) and set1[i] < set2[j]):
            set3.append(set1[i])
            i += 1
        elif i < len(set1) and j < len(set2) and set1[i] == set2[j]:
            set3.append(set1[i])
            i += 1
            j += 1
        elif i == len(set1) or (j < len(set2) and set1[i] > set2[j]):
            set3.append(set2[j])
            j += 1
    
    print ('Объединение: ', set3)
    
    with open('set3.txt', 'w') as f:
        for elem in set3:
            f.write(str(elem))
            f.write(' ')
            


def intersection(filename1, filename2):
# считаем множества из файлов
    with open(filename1, 'r') as f:
        string = f.read()
    set1 = string.split()
    for i in range(len(set1)):
        set1[i] = int(set1[i])
    with open(filename2, 'r') as f:
        string = f.read()
    set2 = string.split()
    for i in range(len(set2)):
        set2[i] = int(set2[i])
        
    i, j = 0, 0
    set3 = []
    while not (i == len(set1) or j == len(set2)):
        if set1[i] < set2[j]:
            i += 1
        elif set1[i] == set2[j]:
            set3.append(set1[i])
            i += 1
            j += 1
        else:
            j += 1
    
    print ('Пересечение: ', set3)
    
    with open('set3.txt', 'w') as f:
        for elem in set3:
            f.write(str(elem))
            f.write(' ')


def difference(filename1, filename2):
    # считаем множества из файлов
    with open(filename1, 'r') as f:
        string = f.read()
    set1 = string.split()
    for i in range(len(set1)):
        set1[i] = int(set1[i])
    with open(filename2, 'r') as f:
        string = f.read()
    set2 = string.split()
    for i in range(len(set2)):
        set2[i] = int(set2[i])
        
    i, j = 0, 0
    set3 = []
    while i < len(set1):
        if j == len(set2) or set1[i] < set2[j]:
            set3.append(set1[i])
            i += 1
        elif set1[i] == set2[j]:
            i += 1
            j += 1
        else:
            j += 1
    
    print ('Разность: ', set3)
    
    with open('set3.txt', 'w') as f:
        for elem in set3:
            f.write(str(elem))
            f.write(' ')


def symmetrical_difference(filename1, filename2):
# считаем множества из файлов
    with open(filename1, 'r') as f:
        string = f.read()
    set1 = string.split()
    for i in range(len(set1)):
        set1[i] = int(set1[i])
    with open(filename2, 'r') as f:
        string = f.read()
    set2 = string.split()
    for i in range(len(set2)):
        set2[i] = int(set2[i])
        
    i, j = 0, 0
    set3 = []
    while i < len(set1) or j < len(set2):
        if j == len(set2) or (i < len(set1) and set1[i] < set2[j]):
            set3.append(set1[i])
            i += 1
        elif set1[i] == set2[j]:
            i += 1
            j += 1
        elif i == len(set1) or (j < len(set2) and set1[i] > set2[j]):
            set3.append(set2[j])
            j += 1
    
    print ('Симметрическая разность: ', set3)
    
    with open('set3.txt', 'w') as f:
        for elem in set3:
            f.write(str(elem))
            f.write(' ')


# main        
combination('set1.txt', 'set2.txt')
intersection('set1.txt', 'set2.txt')
difference('set1.txt', 'set2.txt')
symmetrical_difference('set1.txt', 'set2.txt')
````

# Упражнение 8.  Top down merge sort

Напишите рекурсивный вариант сортировки слиянием. Отталкивайтесь кода C, приведенного [тут](https://en.wikipedia.org/wiki/Merge_sort).

````python
import random


# получим случайный неотсортированный массив с возможно повторяющимися элементами.
A = [random.randint(1, 500) for _ in range(random.randint(1, 25))]


def top_down_merge_sort(a):
    b = a.copy()
    top_down_split_merge(b, 0, len(a) - 1, a)


def top_down_split_merge(b, i_begin, i_end, a):
    if i_end > i_begin:
        i_middle = (i_end + i_begin) // 2
        top_down_split_merge(a, i_begin, i_middle, b)
        top_down_split_merge(a, i_middle + 1, i_end, b)
        top_down_merge(b, i_begin, i_middle, i_end, a)



def top_down_merge(a, i_begin, i_middle, i_end, b):
    i = i_begin
    j = i_middle + 1
    for k in range(i_begin, i_end + 1):
        if i <= i_middle and (j > i_end or a[i] <= a[j]):
            b[k] = a[i]
            i += 1
        else:
            b[k] = a[j]
            j += 1


            
# main        
print('подав на вход массив ', A)
top_down_merge_sort(A)
print('получили на выходе остортированный массив', A)
````
