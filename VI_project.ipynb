```python
Sequence1 = 'Sister'      #str(input())
Sequence2 = 'Monster'       #str(input())

AligmentMatrix = [[0]*(len(Sequence1)+ 1) for _ in range (len(Sequence2)+ 1)]             # хранит список из строк матрицы (тоже списков)

# заполняем дефолтные позиции матрицы
for i in range(len(Sequence1)+1):
    AligmentMatrix[0][i] = i
for j in range(len(Sequence2)+1):
    AligmentMatrix[j][0] = j

# последовательно считаем оставшиеся элементы матрицы
for i in range(1, len(Sequence2)+1):                                                           # Итерация по строчкам вниз
    for j in range(1, len(Sequence1)+1):                                                       # Итерация по столбцам вправо
        if Sequence2[i-1] == Sequence1[j-1]:
            flag = 0
        else:
            flag = 1
        AligmentMatrix[i][j] = min(AligmentMatrix[i][j-1] + 1, AligmentMatrix[i-1][j] + 1,AligmentMatrix[i-1][j-1] + flag )

# печатаем получившуюся матрицу
OutputString = ' \t-\t'
for j in range(len(Sequence1)):
    OutputString += Sequence1[j] + '\t'
print(OutputString)
OutputString = '-\t'
for i in range(len(Sequence2)+1):                                                           # Итерация по строчкам вниз
    for j in range(len(Sequence1)+1):                                                       # Итерация по столбцам вправо
        OutputString += str(AligmentMatrix[i][j]) + '\t'
    print(OutputString)
    if i < len(Sequence2): OutputString = Sequence2[i] + '\t'

# восстанавливаем выровненные последолвательности
outputsequence1 = ''
outputsequence2 = ''

i, j = len(Sequence2), len(Sequence1)
while i > 0 or j > 0:
    #print ('\n', AligmentMatrix[i][j])
    #print(i+1, ' ', j+1)
    if j > 0:
        flag1 = AligmentMatrix[i][j-1] + 1
    if i > 0:
        flag2 = AligmentMatrix[i- 1][j] + 1

    if Sequence2[i - 1] == Sequence1[j - 1]:
        flag = 0
    else:
        flag = 1
    flag3 = AligmentMatrix[i - 1][j - 1] + flag

    minflag = min(flag1, flag2, flag3)

    if minflag == flag1:
        outputsequence1 =  Sequence1[j-1] + outputsequence1
        outputsequence2 = '-' + outputsequence2
        i, j = i, j - 1
    elif minflag == flag2:
        outputsequence1 = '-' + outputsequence1
        outputsequence2 = Sequence2[i - 1] + outputsequence2
        i, j = i - 1, j
    elif minflag == flag3:
        outputsequence1 = Sequence1[j-1]  + outputsequence1
        outputsequence2 = Sequence2[i - 1] + outputsequence2
        i, j = i - 1, j - 1


    flag1, flag2, flag3 = 0, 0, 0

print(outputsequence1)
print(outputsequence2)
```
